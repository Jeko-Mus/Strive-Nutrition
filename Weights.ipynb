```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/chriswmann/datasets/master/500_Person_Gender_Height_Weight_Index.csv")
```

```python
df.head()
```

```python
df.Index.value_counts()
```

```python
plt.scatter(df['Weight'], df['Height'], c=df.Index )
```

### Male info

```python
m_df = df[df['Gender'] == 'Male']
m_df
```

```python
plt.scatter(m_df['Weight'], m_df['Height'], c=m_df.Index )
```

```python
plt.figure(figsize=(15,8))
sns.scatterplot(x='Weight', y='Height', hue='Index', data=m_df ,palette="deep")
plt.show()
```

```python
m_df['Index'].value_counts().plot(kind='bar')

```

### Female info

```python
f_df = df[df['Gender']=='Female']
f_df
```

```python
plt.scatter(f_df['Weight'], f_df['Height'], c=f_df.Index )
```

```python
plt.figure(figsize=(15,8))
sns.scatterplot(x='Weight', y='Height', hue='Index', data=f_df ,palette="deep")
plt.show()
```

```python
f_df['Index'].value_counts().plot(kind='bar')
```

```python
def mean(x):
    sum = 0
    n = len(x)
    for i in x:
        sum = sum + i
        mean = sum/n
    return mean

mean(df.Weight), mean(df.Height)
```
